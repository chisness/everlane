```python
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
import pandas as pd
import numpy as np
driver = webdriver.Firefox()
```

```python
items = []
erroritems = []
```

```python
for url3 in url2:
    name = ''
    cost = 0.0
    everlane_price = 0.0
    retail_price = 0.0
    
    #url3 = 'https://www.everlane.com/products/mens-nylon-duffel-pack-black?collection=mens-backpacks-bags'
    print(url3)
    driver.get(url3)
    html = driver.page_source
    page_content = BeautifulSoup(html)
    #print(page_content.prettify())

    name = page_content.find('h1', class_='product-heading__name')
    #print(name)
    if name is not None:
        name = str(name.find('span', itemprop='name').contents[0])
    else:
        print('name error', url3)
    #print(name)

    cost = page_content.find('div', class_='infographic__true-cost')
    if cost is not None:
        cost = str(cost.find('p', class_='infographic__cost-text infographic__cost-text--price').contents[0])
        cost = float(cost[1:])
    else:
        print('cost error', url3)
    #print(cost)

    everlane_price = page_content.find('span', class_='product-heading__price-value')
    if everlane_price is not None:
        everlane_price = everlane_price.text.strip()
        #everlane_price = retail_price.rsplit(None, 1)[-1]
        everlane_price = float(everlane_price[1:])
    else:
        everlane_price = page_content.find('div', class_='product-page__choose-what-you-pay-price-controls clearfix')
        if everlane_price is not None:
            everlane_price = str(everlane_price.find('button').contents[0])
            everlane_price = float(everlane_price[1:])
        else: 
            print('everlane_price error', url3)
    #print(everlane_price)

    retail_price = page_content.find('div', class_='product-heading__traditional-price')
    if retail_price is not None:
        retail_price = retail_price.text.strip()
        retail_price = retail_price.rsplit(None, 1)[-1]
        retail_price = float(retail_price[1:])
        #print(retail_price)
    else:
        print('retail_price error', url3)

    item = [name, cost, everlane_price, retail_price, url3]
    if item.count(None) == 0:
        items.append(item)
    else:
        erroritems.append(item)
```

```python
d = ['product', 'cost', 'everlane_price', 'retail_price', 'url']
df = pd.DataFrame(columns=d, data=items)
df['everlane_markup'] = df['everlane_price']/df['cost']
df['retail_markup'] = df['retail_price']/df['cost']
df['retail_to_everlane'] = df['retail_price']/df['everlane_price']
df
```
