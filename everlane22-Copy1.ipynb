```python
import pandas as pd 

data = pd.read_csv("everlaned.csv") 
data.head()
```

```python
data.shape
```

```python
data = data.drop(['Unnamed: 0'], axis=1)
```

```python
data = data.drop_duplicates(subset='product', keep="first")
```

```python
data.shape
```

```python
data = data.sort_values(by=['everlane_markup'])
```

```python
data.reset_index()
```

```python
data.hist(column=['everlane_markup', 'retail_markup', 'retail_to_everlane'],figsize=(20,10))
```

```python
#best value 10 items (lowest everlane markup)
bv = data.nsmallest(10, 'everlane_markup').reset_index()
bv
```

```python
#worst value 5 items
wv = data.nlargest(10, 'everlane_markup').reset_index()
wv
```

```python
#best value 5 items not on sale
data_ns = data[data['sale']!=True]
bvns = data_ns.nsmallest(5, 'everlane_markup').reset_index()
bvns
```

```python
data.to_csv('everlaned2.csv')
```
